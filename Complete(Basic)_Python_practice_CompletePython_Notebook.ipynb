### 5. Variable type casting

The process of converting given variable data type (example - integer, string, float, etc.) to another data type is called type conversion.

Python has two types of type conversion.

- `Implicit` Type Conversion
     - Python automatically converts one data type to another data type. This process doesn't need any user involvement.

- `Explicit` Type Conversion
     - Python need user involvement to convert the variable data type into certain data type in order to the required operation.

`Syntax : <required_datatype>(expression)`

#### Order of precedence

Arithmetic Operators in Python follow a basic order of precedence.

When more than one operator is used, they are executed according to below order.

1. `()`	Parentheses

2. `**`	Exponent

3. `%`, `*`, `/`, `//`	Modulos, Multiplication, Division and Floor division

4. `+`,`-`	Addition and Subtraction


Note : The operator listed at the top of the table will be executed first.

Given a matrix of numbers, return a new matrix by multiplying the original number with the place value. 10s for 1st row, 100s for second row and so on

````python
nums = [[1,2],
        [3,4]]

m = 1
for i in range(len(nums)):
  m = m*10
  for j in range(len(nums[i])):
    nums[i][j] = nums[i][j]*m
print(nums)
````

````python
words = ["data","science","machine","learning","good","morning"]

words_dict = {}
for word in words:
  if len(word)>5:
    words_dict[word] = len(word)
  else:
    words_dict[word] = "short"

print(words_dict)
````

Recursion :- The term recursion can be defined as the process of defining something in term of itself. In simple word, it is a process in which a function call itself directly or inderectly






````python
def simple_recursion(n):
  if n == 1:
    return  1
  else:
    return n + simple_recursion(n-1)
simple_recursion(5)
````

````python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    return (fibonacci(n-1) + fibonacci(n-2))

fibonacci(5)
````

````python
def factorial(n):
  if n == 1:
    return 1
  if n < 0:
    print("Factorial does not exist")
  elif n == 0:
    return "Factorial of 0 is 1"
  else:
    return n*factorial(n-1)

factorial(5)
````
